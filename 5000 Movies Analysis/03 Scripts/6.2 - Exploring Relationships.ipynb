## Create a new column as group the country into region to help analyzing in furture step

```python
region = []
for value in movie['country']:
    if value in ('AF','KR','IN','HK','MY','PH','AE','SG','TH','FJ','CN','JP','IL','BT','LB','KG','ID','PK','IR'):
        region.append('Asia')
    elif value in ('DO','PE','BR','MX','AW','AR','PA','BS','US','JM','BO','EC','GY','GP'):
        region.append('America')
    elif value in ('CM','KE','ZA','LY','MA','PZ'):
        region.append('Africa')
    elif value in ('SI','NL','BE','NO','IE','FI','IS','DK','RU','BG','CH','MT','BA','LU','RO','HU','AT','GR','SE','TR','GB','CZ','NZ','DE','FR','IT','ES','CY','PL'):
        region.append('Europe')
    elif value =='AU':
        region.append('Oceania')
    else:
        region.append('Unknown')
```

```python
movie['region'] = region
```

#### Question created from previous task:
How can we define a movie as popularity? Are there any correlation actors among the observations?

##### Step 1: Finding out the variables that would effect the movie to be popularity

```python
# Create a subset that only contains the possible variables that related with popularity
sub = movie[['popularity','budget','revenue','runtime','vote_count','region']]
sub
```

##### Step 2: Create a heatmap by using seaborn to show the correlations

```python
f, ax = plt.subplots(figsize=(10,6))

sns.heatmap(sub.corr(), 
            annot=True, 
            cbar=True,
            ax=ax,
            cmap='Blues',
           linewidth=0.1)
plt.savefig(os.path.join(path,'correlation_heatmap.png'))
plt.show()
```

The coefficients in between popularity and each variable are:

1: revenue has a coefficient 0.57 with popularity as considered a strong relationship with each other, the higher revenue, the more popularity gains.

2: budget has a cofficient 0.46 with popularity that shows a moderate relationship with each other, as the higher budget, may cause a high popularity normally. 

3: runtime has a coefficient 0.22 with popularity as considered a week relationship, so the longer runtime may be not considered as a populary movie.

4: vote_count has the strongest coefficient 0.78 with popularity which means the more votes gains, the more popularity is for sure.

#### Create a scatter plot between popularity and vote_count

```python
sns.set_style('ticks')
sns.lmplot(x = 'popularity',
           y = 'vote_count', 
           markers = ["D"],
           data = sub,
           aspect = 1.5
          )
sns.despine(offset=20, trim=True)
plt.savefig(os.path.join(path,'scatter.png'))
plt.show()
```

#### Create a categorical plot and interpret the results

##### Step 1: Using histogram to visualize the distribution of the variables

```python
sns.set(rc={'figure.figsize':(8,4)})
sns.set_style('ticks')
sns.histplot(movie['runtime'], bins=35, kde=True, color='g')
sns.despine()
plt.title('Runtime Distribution')
plt.savefig(os.path.join(path,'hist_runtime.png'))
plt.show()
```

##### Step 2: Create a duration_flag based on the visualization showed above

```python
duration = []
for value in movie['runtime']:
    if value <= 90:
        duration.append('Short duration')
    elif value > 90 and value <= 120:
        duration.append('Middle duration')
    else:
        duration.append('Long duration')
```

```python
movie['duration_flag'] = duration
```

```python
movie['duration_flag'].value_counts()
```

```python
sns.histplot(x = 'vote_average', bins=10, kde=True, data=movie)
sns.despine()
plt.show()
```

```python
# Create recommend_flag based on the distribution showed above
movie.loc[movie['vote_average'] <= 6, 'recommend_flag'] = 'Not Recommend'
movie.loc[(movie['vote_average'] > 6) & (movie['vote_average'] <= 7.5), 'recommend_flag'] = 'Recommend'
movie.loc[movie['vote_average'] > 7.5, 'recommend_flag'] = 'Strong Recommend'
```

```python
movie['recommend_flag'].value_counts()
```

```python
sns.set(rc={'figure.figsize':(10,4)})
sns.set_style('ticks')
sns.histplot(x='budget', 
             bins=35, 
             kde=True,
             data=movie)
sns.despine()
plt.ticklabel_format(axis='x', style='plain')
plt.show()
```

```python
# Create a budget spending flag to explory more
spent = []
for value in movie['budget']:
    if value < 10000000:
        spent.append('Under 10M')
    elif value > 10000000 and value <= 50000000:
        spent.append('10M to 50M')
    elif value > 50000000 and value <= 100000000:
        spent.append('50M to 100M')
    else:
        spent.append('Above 100M')
```

```python
movie['budget_flag'] = spent
```

```python
movie.to_csv(os.path.join(path,'movie_derived.csv'))
```
